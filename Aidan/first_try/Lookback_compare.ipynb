```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import glob
```

```python
paths = glob.glob("./Lookback_sweep/*.txt")
```

```python
plt.figure(figsize=(10,10))
for path in np.sort(paths):
    lookback = path.split("_")[2].split(".")[0]
    x = pd.Series(np.loadtxt(path,dtype="float")).cumsum().index.values
    returns = pd.Series(np.loadtxt(path,dtype="float")).cumsum()
    plt.plot(x,returns, label=lookback)
plt.axvline(1000, linestyle="--")
plt.legend(loc = (1,0.5))
plt.title("Lookback of 160 Peforms well")
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(20,10), sharey=True)
for path in np.sort(paths):
    lookback = path.split("_")[2].split(".")[0]
    x = pd.Series(np.loadtxt(path,dtype="float")).cumsum().index.values
    returns = pd.Series(np.loadtxt(path,dtype="float")).cumsum()
    if int(lookback) > 100:
        axs[1].plot(x,returns, label=lookback)
    else:
        axs[0].plot(x,returns, label=lookback)
axs[0].axvline(1000, linestyle="--")
axs[1].axvline(1000, linestyle="--")

axs[0].legend(loc = (1.05,0.5))
axs[1].legend(loc = (1.05,0.5))

axs[0].set_title("<=100 lookback")
axs[1].set_title(">100 Lookback")

plt.suptitle("Larger lookback state space performs better", fontsize=20)
plt.savefig("Lookback.png")
plt.show()
```
